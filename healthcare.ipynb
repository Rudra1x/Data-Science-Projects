```python
def parse(x):
    y = x.split()
    t = y[1][:8]
    z=y[0] + " " + t
    d = datetime.strptime(z, "%Y-%m-%d %H:%M:%S")
    return d
```

```python
df = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/data-science-in-health-care-basic-statistical-analysis/COVID_19.xlsx',
    'Sheet1',
    na_values="NaN",
    parse_dates=['Date time'],
    index_col=0,
    date_parser=parse)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ['Gender'])
```

```python
d = {'No': False,'Yes': True}
c = 'Do you vaccinated influenza?'
df.loc[:,c] = df[c].map(d)
```

```python
df.head()
```

```python
df.info()
```

```python
c = 'Age'
df.loc[:,c] = df[c].astype('category')
```

```python
df['Age']
```

```python
for c in df.columns[1:-1]:
    df.loc[:, c] = df[c].apply(lambda x : str(x) if str(x).find('(') == -1 else str(x)[:str(x).find('(')]).astype('category')
```

## Visualize Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check for duplicate indices
print("Number of duplicate indices:", df.index.duplicated().sum())
print("DataFrame shape:", df.shape)
print("Index type:", type(df.index))

# For plotting that doesn't need datetime index, reset the index
df_clean = df.reset_index(drop=True)
print("After resetting index - DataFrame shape:", df_clean.shape)

# For time series analysis, keep datetime index but handle duplicates differently
df_timeseries = df[~df.index.duplicated(keep='first')]
print("After removing duplicate datetime indices:", df_timeseries.shape)
print("Timeseries index type:", type(df_timeseries.index))
```

```python
# Use the cleaned dataframe with reset index
sns.countplot(x='Age', hue='Gender', data=df_clean)
plt.title('Count of Age Groups by Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Use df_timeseries for time-based operations like resample
# First, let's check what numeric columns we have for summing
print("Numeric columns:")
numeric_cols = df_timeseries.select_dtypes(include=['number']).columns
print(numeric_cols.tolist())

# If we have numeric columns, resample and plot
if len(numeric_cols) > 0:
    # Resample by day and sum numeric columns
    daily_data = df_timeseries[numeric_cols].resample('1D').sum()
    daily_data.plot(figsize=(12, 6))
    plt.title('Daily Sum of Numeric Variables')
    plt.xlabel('Date')
    plt.ylabel('Sum')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns found for resampling")
    
    # Alternative: Count records per day
    daily_counts = df_timeseries.resample('1D').size()
    daily_counts.plot(figsize=(12, 6))
    plt.title('Daily Count of Records')
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
